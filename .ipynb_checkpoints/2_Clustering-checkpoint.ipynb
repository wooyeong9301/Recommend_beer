## 클러스터링

```python
import warnings
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
warnings.filterwarnings('ignore')
pd.set_option('display.max_rows', 100)
```

```python
data = pd.read_csv('beer_final.csv')
data.reset_index(drop=True, inplace=True)
```

### 맥주 별 평점 계산

```python
# 새로운 dataframe 생성
beer_avg = pd.DataFrame(data=[], columns=['맥주', 'Aroma', 'Appearance', 'Flavor', 'Mouthfeel', 'Overall'])

beer_avg['맥주'] = data['맥주이름'].drop_duplicates(keep='first')

for i in range(len(beer_avg)):
    # 각각의 맥주에 대해
    beer = beer_avg['맥주'].iloc[i]
    # 각 평가 요소의 평균 입력
    for _ in beer_avg.columns[1:]:
        tmp = round(data[data['맥주이름']==beer][_].astype(int).mean(), 2)
        beer_avg[_].iloc[i] = tmp
```

```python
beer_avg.reset_index(drop=True, inplace=True)
beer_avg
```

### 클러스터링

```python
# 실수형 컬럼들을 뽑아 표준화
beer_names = beer_avg[['맥주']]
beer_values = beer_avg[['Aroma', 'Appearance', 'Flavor', 'Mouthfeel', 'Overall']]

scaler = MinMaxScaler(feature_range=(0,5))
scaler.fit(beer_values)

scaled = scaler.transform(beer_values)
scaled
```

```python
# elbow point
def elbow(X):
    sse = []
    for i in range(1,11):
        km = KMeans(n_clusters=i)
        km.fit(X)
        sse.append(km.inertia_)
    plt.figure(1, figsize=(6,6))
    plt.plot(range(1,11), sse, 'o')
    plt.plot(range(1,11), sse, '-', alpha=0.5)
    plt.xlabel('Number of Clusters')
    plt.ylabel('Inertia')
    
elbow(scaled)
```

```python
km = KMeans(n_clusters=4).fit(scaled)
km.cluster_centers_
```

```python
# 맥주 별 군집 예측
predict = pd.DataFrame(km.predict(scaled))
predict.columns = ['Cluster']
predict.head()
```

```python
scaled_features = pd.DataFrame(data=scaled, columns=['Aroma', 'Appearance', 'Flavor', 'Mouthfeel', 'Overall'])
scaled_features
```

```python
beer_names.reset_index(inplace=True, drop=True)
result = pd.concat([scaled_features, beer_names], axis=1).reset_index(drop=True)
```

```python
result = pd.concat([result, predict], axis=1).reset_index(drop=True)
```

```python
result
```

```python
# 각 클러스터 군집들의 features는 어떤지?
c_result = km.cluster_centers_
c_result = pd.DataFrame(data=c_result)
c_result.columns = ['Aroma', 'Appearance', 'Flavor','Mouthfeel','Overall']
c_result.sort_values(by='Overall', inplace=True)
c_result
```

### 시각화

```python
def show_cluster(cluster, data, name):
    categories = c_result.columns
    categories = [*categories, categories[0]]

    new_c = pd.concat([c_result, c_result[['Aroma']]], axis=1)

    target = data[data['맥주']==name]
    cluster = target['Cluster'].iloc[0]
    target = target[categories]

    ll = np.linspace(start=0, stop=2*np.pi, num=len(categories))
    plt.figure(figsize=(8,8))
    ax = plt.subplot(polar=True)

    plt.xticks(ll, labels=categories, fontsize=13)
    ax.plot(ll, new_c.iloc[0], label='c1', linestyle='--', color='red')
    ax.fill(ll, new_c.iloc[0], color='red', alpha=0.8)
    ax.plot(ll, new_c.iloc[1], label='c1', linestyle='--', color='salmon')
    ax.fill(ll, new_c.iloc[1], color='salmon', alpha=0.6)
    ax.plot(ll, new_c.iloc[2], label='c1', linestyle='--', color='skyblue')
    ax.fill(ll, new_c.iloc[2], color='skyblue', alpha=0.4)
    ax.plot(ll, new_c.iloc[3], label='c1', linestyle='--', color='blue')
    ax.fill(ll, new_c.iloc[3], color='blue', alpha=0.2)

    ax.plot(ll, target.iloc[0], label='target', linestyle='solid', color='black')
    ax.fill(ll, target.iloc[0])

    return print('cluster :', cluster), plt.show()

show_cluster(c_result, result, 'FiLite')
```

```python
def new_cosine(u, v):
    m = u.shape[0] 
    udotv = 0
    u_norm = 0
    v_norm = 0
    for i in range(m):
        if (np.isnan(u[i])) or (np.isnan(v[i])): # Logical OR operator 참고
            continue
            
        udotv += u[i] * v[i]
        u_norm += u[i] * u[i]
        v_norm += v[i] * v[i]

    u_norm = np.sqrt(u_norm)
    v_norm = np.sqrt(v_norm)
    
    if (u_norm == 0) or (v_norm == 0):
        ratio = 1.0
    else:
        ratio = udotv / (u_norm * v_norm)
    return ratio
```
