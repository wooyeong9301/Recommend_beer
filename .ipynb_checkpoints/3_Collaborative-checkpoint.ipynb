# Collaborative Filtering
협업 필터링은 최근접 이웃 협업 필터링과 잠재 요인 협업 필터링으로 나뉘고,    
최근접 이웃 협업 필터링은 다시 유저 기반과 아이템 기반으로 나뉜다.   
맥주 추천의 경우 유저의 취향은 너무 다양한 데 비해, 리뷰를 남기는 맥주의 수는 한정적이므로 아이템 기반 추천을 사용해본다.

```python
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('beer_final.csv', index_col=0)
```

### 데이터 정제
최소 n개 이상의 리뷰를 가진 맥주, 유저들의 데이터셋 구성

```python
data
```

```python
n_ids = data['유저'].value_counts() >= 10
n_ids = n_ids[n_ids].index.to_list()
data = data[data['유저'].isin(n_ids)]
print(len(data))

n_rvws = data['맥주이름'].value_counts() >= 10
n_rvws = n_rvws[n_rvws].index.to_list()
data = data[data['맥주이름'].isin(n_rvws)]
print(len(data))
```
