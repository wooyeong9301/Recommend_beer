# Collaborative Filtering
협업 필터링은 최근접 이웃 협업 필터링과 잠재 요인 협업 필터링으로 나뉘고,    
최근접 이웃 협업 필터링은 다시 유저 기반과 아이템 기반으로 나뉜다.   
맥주 추천의 경우 유저의 취향은 너무 다양한 데 비해, 리뷰를 남기는 맥주의 수는 한정적이므로 아이템 기반 추천을 사용해본다.

```python
import pandas as pd
import numpy as np
import pickle

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('beer_final.csv', index_col=0)
```

### 데이터 정제
최소 n개 이상의 리뷰를 가진 맥주, 유저들의 데이터셋 구성

```python
# n=10
n_ids = data['유저'].value_counts() >= 10
n_ids = n_ids[n_ids].index.to_list()
data = data[data['유저'].isin(n_ids)]
print(len(data))

n_rvws = data['맥주이름'].value_counts() >= 10
n_rvws = n_rvws[n_rvws].index.to_list()
data = data[data['맥주이름'].isin(n_rvws)]
data = data.reset_index(drop=True)
print(len(data))
```

### EDA

```python
from scipy.stats import normaltest
```

```python
# outlier check
def check_outlier(df):
    features = df.columns[-5:]
    outlier_idxs = []
    for ft in features:
        q3 = data[ft].quantile(0.75)
        q1 = data[ft].quantile(0.25)
        iqr = q3 - q1
        condition1 = data[ft] > q3+1.5*iqr
        condition2 = data[ft] < q1-1.5*iqr
        outlier_idxs.extend(data[condition2].index)
        outlier_idxs.extend(data[condition2].index)
    
    new_df = df.drop(outlier_idxs, axis=0)
    new_df.reset_index(drop=True, inplace=True)
    if len(df) == len(new_df):
        print('There is no outlier!')
    else:
        print('There is(are) {} outlier(s)!'.format(len(df)-len(new_df)))
    return new_df
```

```python
data = check_outlier(data)
```

### 맥주 평점을 이용하여 협업 필터링
평점의 유사도를 기반으로 Memory-Based Collaborative Filtering 중 User-Based Filtering 진행   
: 타겟과 유사한 사용자가 좋아한 아이템을 추천

```python
# user-item matrix를 이용해 cosine similarity 계산
dd = data.copy()

# pivot table로 user-item matrix 생성
ratings_matrix = dd.pivot_table('평점', index='유저', columns='맥주이름')
```

```python
with open('ratings_matrix.pkl', 'wb') as f:
    pickle.dump(ratings_matrix, f)
```

```python
with open('ratings_matrix.pkl', 'rb') as f:
    ratings_matrix = pickle.load(f)
```

```python
ratings_matrix
```

### 1. null값을 0으로 처리

```python
rm_fill_zero = ratings_matrix.fillna(0)
```

```python
# user-based cosine similarity
user_sim = cosine_similarity(rm_fill_zero, rm_fill_zero)
user_sim = pd.DataFrame(user_sim, index=rm_fill_zero.index, columns=rm_fill_zero.index)
print(user_sim.shape)

for _ in user_sim.index[:3]:
    print('Most similar user for {}'.format(_))
    print(user_sim[_].sort_values(ascending=False)[1:6])
    
def find_sim_user(user, k):
    print('Most similar {} user for {}'.format(k, user))
    return user_sim[user].sort_values(ascending=False)[1:k+1]
```

```python
# item-based cosine similarity
item_sim = cosine_similarity(rm_fill_zero.T, rm_fill_zero.T)
item_sim = pd.DataFrame(item_sim, index=rm_fill_zero.columns, columns=rm_fill_zero.columns)
print(item_sim.shape)

for _ in item_sim.index[:3]:
    print('Most similar item for {}'.format(_))
    print(item_sim[_].sort_values(ascending=False)[1:6])
    
def find_sim_item(item, k):
    print('Most similar {} item for {}'.format(k, item))
    return item_sim[item].sort_values(ascending=False)[1:k+1]

find_sim_item('San Miguel',10)
```

### 개인화된 맥주 추천
사용자의 평점에 맥주 간의 유사도를 가중치로 계산한다.  
사용자 A의 맥주 a에 대한 예측 평점은, rm(A) x item(a) / sum(item(a))

```python
# 평점 매트릭스 dot 맥주 유사도 매트릭스
def predict_rating(ratings_matrix, item_sim):
    rating_pred = ratings_matrix.dot(item_sim)/item_sim.sum(axis=1)
    return rating_pred

ratings_pred = predict_rating(rm_fill_zero, item_sim)
```

```python
# 개인에 대한 추천
def personal_beer(user, k):
    print('Top-{} recommendation for user {}'.format(k, user))
    
    tmp = []
    for _ in rm_fill_zero.columns:
        if np.isnan(ratings_matrix.loc[user][_]):
            tmp.append(_)
        else:
            continue
    
    recommendation = ratings_pred[tmp]
    
    return recommendation.loc[user].sort_values(ascending=False)[1:k+1]
```

```python
personal_beer(69,7)
```

NaN을 으로 대체하였기 때문에 예측된 평점이 실제 평점의 평균보다 훨씬 낮다.
